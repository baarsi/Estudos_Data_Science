Desafio: Colorir o seu estado com um tom diferente. Colorir os outros estados de acordo com gasto maior ou menor.


**matplotlib.patches** é um módulo do Matplotlib que fornece uma série de objetos gráficos para desenhar formas geométricas, como retângulos, círculos, elipses, polígonos, setas e mais. Essas formas podem ser usadas para adicionar elementos gráficos personalizáveis ao seu gráfico, como áreas de destaque ou anotações.

```python
import matplotlib.patches as mpatches

# Ordenando a tabela de comparação em ordem decrescente
tabela_de_comparacao = tabela_de_comparacao.sort_values(ascending=False)

# Selecionando o valor-alvo e o índice do estado-alvo
target_index2 = "33 Rio de Janeiro"
target_value = tabela_de_comparacao.loc[target_index2]

# Definindo as cores com base na lógica
cores = [
    "green" if index == target_index2 else
    "blue" if value > target_value else
    "purple"
    for index, value in tabela_de_comparacao.items()
]

# Plotando o gráfico de barras
tabela_de_comparacao.plot(kind="bar", figsize=(12, 6), color=cores)

# Adicionando uma linha horizontal na altura do valor-alvo
plt.axhline(y=target_value, color='red', linestyle='--')

# Configurações de eixos e título
plt.xlabel("Estados", fontsize=12)
plt.ylabel("Valores", fontsize=12)
plt.title("Comparação de Gastos por Unidade da Federação", fontsize=14)

# Adicionando a grade no eixo Y
plt.grid(axis="y", linestyle="--", alpha=0.7) # Aprimorando a grade

# Criando de patches para a legenda
estado_alvo_patch = mpatches.Patch(color='green', label=f"Estado-alvo ({target_index2})")
maior_gasto_patch = mpatches.Patch(color='blue', label="Gasto maior que o alvo")
menor_gasto_patch = mpatches.Patch(color='purple', label="Gasto menor que o alvo")
linha_alvo_patch = mpatches.Patch(color='red', label="Linha de referência (gasto do alvo)", linestyle='--')

# Adicionando as legendas
plt.legend(handles=[estado_alvo_patch, maior_gasto_patch, menor_gasto_patch, linha_alvo_patch], loc='best')

# Ajustando o layout e exibindo o gráfico
plt.tight_layout()
plt.show()
```

O **tight_layout()** ajusta o posicionamento dos elementos automaticamente com base no conteúdo, de forma que haja espaço suficiente entre os elementos e nada seja cortado ou se sobreponha.

**mpatches.Patch:** Esta função é utilizada para criar um objeto do tipo "patch", que pode ser um retângulo, círculo, ou outras formas. Aqui está sendo usado para criar um retângulo simples (também conhecido como "patch" de cor única).

**mpatches:** É um alias para matplotlib.patches, que é um módulo da biblioteca Matplotlib utilizado para criar formas geométricas.

**color=:** Define a cor do patch.

**label="Gasto maior que o alvo":** A etiqueta (label) que será associada a esse patch na legenda do gráfico. Esse texto será exibido na legenda, explicando que a cor azul representa valores em que o gasto é maior que o alvo.

**plt.axhline():** Esta função é usada para desenhar uma linha horizontal no gráfico, em um valor específico do eixo Y. Ela é útil para marcar um valor de referência ou uma linha de limite no gráfico.

**y=target_value:** O parâmetro y define a posição da linha no eixo vertical (eixo Y). Nesse caso, a linha será desenhada na altura correspondente ao valor de target_value.

**color='red':** O parâmetro color define a cor da linha. Neste caso, a linha será vermelha ('red').

**linestyle='--':** O parâmetro linestyle define o estilo da linha. Neste caso, '--' indica que a linha será tracejada (linha pontilhada).
