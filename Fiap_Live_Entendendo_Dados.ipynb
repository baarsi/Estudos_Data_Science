```python
# Restringindo os índices em um novo DataFrame
df_idhm = idhm[['Nome da Unidade da Federação', 'Município', 'IDHM', 'IDHM Educação', 'IDHM Longevidade', 'IDHM Renda']]
df_idhm.head()
```

```python
# Renomeando as colunas do DataFrame
df_idhm.columns = ['UF', 'municipio', 'idhm', 'idhm_educacao', 'idhm_longevidade', 'idhm_renda']
df_idhm.head()
```

**Análise inicial**

0 - 0,499 - Baixo Desenvolvimento Humano<br/>
0,5 - 0,799 - Médio Desenvolvimento Humano<br/>
0,8 - 1 - Alto Desenvolvimento Humano

```python
# Classificando os dados
def idhm_class(df):
    if df['idhm']>=0.8:
        return 'Alto Desenvolvimento Humano'
    elif df['idhm']<0.499:
        return 'Baixo Desenvolvimento Humano'
    else:
        return 'Médio Desenvolvimento Humano'
```

```python
# Usando '.apply()' para aplicar a função em cada elemento, linha ou coluna
df_idhm['classe'] = df_idhm.apply(idhm_class, axis=1)
df_idhm
```

```python
# Exibindo as 20 primeiras linhas do DataFrame
df_idhm.head(20)
```

```python
# criando uma "grade de gráficos" (grid)
# cada variável é plotada contra cada outra variável do conjunto de dados, formando pares
sns.pairplot(df_idhm, hue = 'classe')

# obs: é possível observar uma aparente relação entre renda e escolaridade
```

```python
# Realizando análise por Estado (UF)
df_idhm_estado = pd.DataFrame(df_idhm.groupby('UF')['idhm'].mean().round(2)).reset_index()
df_idhm_estado
```

```python
# '.groupby()' para agrupar dados em um DataFrame com base nos valores de uma ou mais colunas
# ideal para dividir o DataFrame em grupos e aplicar operações sobre esses grupos
# reset_index() é usado para redefinir o índice de um DataFrame
```

```python
df_idhm_estado['classe'] = df_idhm_estado.apply(idhm_class, axis=1)
df_idhm_estado
```

```python
df_idhm_estado.info()
```

```python
df_idhm_estado.describe()
```

```python
# Usando SEABORN
# Criando gráficos de caixa (box plot)
sns.boxplot(y=df_idhm_estado['idhm'], data=df_idhm_estado)

# Adicionar título e rótulos dos eixos
plt.title('Boxplot dos Valores por Classificação')
plt.xlabel('IDHM')
plt.ylabel('Valores')
```

```python
# Criando um gráfico de barras
# Tamanho da Figura
plt.figure(figsize=(12,6))

# Paleta de Cores
Paleta_Cores = sns.color_palette('mako', len(df_idhm_estado) )
sns.set_style('darkgrid')

# Plotagem
plt.bar(df_idhm_estado['UF'],
        df_idhm_estado['idhm'],
        width=0.9,
        color = Paleta_Cores
        )

# Titulo
plt.title('UF x IDHM', loc = 'left', fontsize = 18)

# Labels
plt.xlabel('UF', fontsize = 14)
plt.ylabel('IDH com a média dos Estados', fontsize = 14)
plt.xticks(rotation = 90);
```

```python
# Criando um gráfico de barras usando o Plotly
fig = px.bar(df_idhm_estado, x='UF', y='idhm', text='idhm', color = 'classe', color_discrete_sequence=['orange','green'])


# Adicionando título e rótulos dos eixos
fig.update_layout(title='Gráfico de Barras',
                  xaxis_title='UF',
                  yaxis_title='idhm')
fig.show()
```

```python
# Usando o Plotly
fig = px.treemap(df_idhm_estado, path=['UF'], values='idhm', title='TreeMap')

# Exibindo o treemap
fig.update_traces(root_color="lightgrey")
fig.show()
```
