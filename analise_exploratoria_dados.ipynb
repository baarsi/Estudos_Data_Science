```python
import pandas as pd
```

```python
dados = pd.read_csv("datasus_ano_mes_atendimento.csv", encoding="ISO-8859-1", sep=";", skiprows=3, skipfooter=12, thousands=".", decimal=",", engine="python")
```

```python
dados2 = pd.read_csv("datasus_ano_mes_processamento.csv", encoding="ISO-8859-1", sep=";", skiprows=3, skipfooter=12, thousands=".", decimal=",", engine="python")
```

```python
# Selecionando a coluna mais recente (última) do Dataframe 'dados2' - retorna uma Serie
dados2['2016/Dez']
```

```python
# Retorna a média dos valores contidos na coluna '2016/Dez'
dados2['2016/Dez'].mean()
```

```python
# Gerando um gráfico com o abscissa "Unidade da Federação" e ordenada "2016/Dez" - usando méthodo '.plot()'
dados2.plot(x="Unidade da Federação", y="2016/Dez")
```

```python
# Importando 'matplotlib.pyplot' e atribuindo um alias ('plt')
import matplotlib.pyplot as plt

# Importando 'matplotlib.ticker' e atribuindo um alias ('ticker')
import matplotlib.ticker as ticker
```

```python
# Estilizando o gráfico para o tipo "barras"
axis = dados2.plot(x="Unidade da Federação", y="2016/Dez", kind="bar", figsize=(9,6))

# Configurando o eixo y (formatação numérica)
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por unidade da federação")
plt.show()
```

```python
#Desafio FIAP

# Alterando a escala do gráfico (25 x 6)
axis = dados2.plot(x="Unidade da Federação", y="2016/Dez", kind="bar", figsize=(25,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor por unidade da federação")

# Aplicando uma rotação do tick (eixo x) de 45°
plt.xticks(rotation=45)
plt.show()
```

```python
# Alterando a escala do gráfico (50 x 6)
axis = dados2.plot(x="Unidade da Federação", y="2016/Dez", kind="bar", figsize=(50,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor por unidade da federação")

# Aplicando uma rotação do tick (eixo x) de 0°
plt.xticks(rotation=0)
plt.show()
```
