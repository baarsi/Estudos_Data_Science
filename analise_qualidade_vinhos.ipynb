```python
import pandas as pd

data = pd.read_csv("winequalityN.csv")
```

```python
data.head(5)
```

```python
data.info()
```

```python
data.dropna() # removendo valores nulos
```

```python
data.drop_duplicates() # removendo linhas duplicadas
```

4.a. Distribuição da acidez volátil por qualidade do vinho

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='quality', y='volatile acidity', hue='quality', legend=False, palette="pastel")

plt.title('Volatile Acidity x Wine Quality', fontsize=16)
plt.xlabel('Quality', fontsize=12)
plt.ylabel('Acidity', fontsize=12)

plt.show()
```

4.b. Distribuição do teor alcóolico por qualidade do vinho

```python
aggregated_data = data.groupby('quality', as_index=False).mean(numeric_only=True)

sns.barplot(data=aggregated_data, x='quality', y='alcohol', palette="deep")
plt.title("Alcohol x Quality")
plt.show()
```

4.c. Relação entre acidez volátil e teor alcóolico

```python
# Gráfico de dispersão com linha de regressão
sns.regplot(data=data, x='volatile acidity', y='alcohol', scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
plt.title("Volatile Acidity x Alcohol")
plt.show()
```

```python
# Gráfico de linha
data = data.sort_values('volatile acidity')
sns.lineplot(data=data, x='volatile acidity', y='alcohol', hue='quality', marker='o')
```

```python
# Gráfico de Hexbin - Útil para grandes conjuntos de dados, agregando pontos em hexágonos.
plt.hexbin(data['volatile acidity'], data['alcohol'], gridsize=30, cmap='coolwarm', alpha=0.8)
plt.colorbar(label='Contagem')
plt.xlabel('Volatile Acidity')
plt.ylabel('Alcohol')
plt.title("Hexbin: Volatile Acidity x Alcohol")
plt.show()
```

```python
# Gráfico de Densidade (KDE) - Visualiza as distribuições conjuntas de duas variáveis como áreas de densidade
sns.kdeplot(data=data, x='volatile acidity', y='alcohol', cmap='viridis', fill=True)
plt.title("KDE Plot: Volatile Acidity x Alcohol")
plt.show()
```

4.d. Relação entre teor alcóolico e qualidade do vinho

```python
sns.regplot(data=data, x='alcohol', y='quality', scatter_kws={'alpha': 0.6}, line_kws={'color': 'blue'})
plt.title("Alcohol x Quality")
plt.show()
```

```python
sns.kdeplot(data=data, x='alcohol', y='quality', cmap='viridis', fill=True)
plt.title("KDE Plot: Alcohol x Quality")
plt.show()
```

5.b. Qual a relação entre teor alcóolico e qualidade do vinho?

```python
sns.regplot(data=data, x='alcohol', y='quality', scatter_kws={'alpha': 0.6}, line_kws={'color': 'blue'})
plt.title("Alcohol x Quality")
plt.show()
```

O gráfico sugere uma relação positiva entre o teor alcoólico e a qualidade do vinho, conforme indicado pela linha de tendência ascendente. Implicando que, em média, vinhos com maior teor alcoólico tendem a ter uma qualidade melhor, segundo a escala representada. No entanto, também é possível observar uma significativa dispersão nos dados, com vinhos de qualidade similar apresentando diferentes teores alcoólicos. Isso indica que, embora o teor alcoólico possa ser um fator influente, ele não é o único determinante da qualidade do vinho. Portanto, fatores adicionais devem ser considerados para avaliar o padrão de qualidade.

5.c. Há alguma relação entre acidez volátil e teor acóolico?

```python
sns.kdeplot(data=data, x='volatile acidity', y='alcohol', cmap='viridis', fill=True)
plt.title("KDE Plot: Volatile Acidity x Alcohol")
plt.show()
```

À medida que a acidez volátil aumenta (acima de 0,6), observa-se uma menor densidade de dados, indicando que vinhos com maior acidez volátil são menos frequentes. Por outro lado, para níveis mais altos de teor alcoólico (acima de 12), a acidez volátil tende a estar em níveis mais baixos, sugerindo uma possível relação inversa entre esses dois fatores. No entanto, a distribuição dos dados não aponta uma correlação linear clara.

4.d. Há alguma relação entre acidez volátil e qualidade do vinho?

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='quality', y='volatile acidity', hue='quality', legend=False, palette="pastel")

plt.title('Volatile Acidity x Wine Quality', fontsize=16)
plt.xlabel('Quality', fontsize=12)
plt.ylabel('Acidity', fontsize=12)

plt.show()
```

É possível inferir que existe uma corrlação inversa entre o nível de acidez e a qualidade do vinho, representando que um vinho com maior qualidade tende a possuir um nível menor de acidez.
